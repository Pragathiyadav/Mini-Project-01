# 1.    Univariate Analysis

1. Bar Plot:

A bar plot is used to visualize the distribution of categorical data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
data = pd.read_csv("C:/Users/DELL/Downloads/archive (9)/Walmart Data Analysis and Forcasting.csv")

# Create a bar plot for the 'Store' column
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Store')
plt.title('Bar Plot of Store Distribution')
plt.xticks(rotation=90)
plt.show()
```

2. Box Plot:

A box plot visualizes the distribution of numerical data, showing quartiles, potential outliers, and the overall spread of the data.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Weekly_Sales')
plt.title('Box Plot of Weekly Sales')
plt.show()
```

3. FacetGrid (for Univariate Analysis):

A FacetGrid is useful for creating multiple plots for different categories within a single variable.

```python
g = sns.FacetGrid(data, col='Holiday_Flag')
g.map(plt.hist, 'Weekly_Sales', bins=20)
```

4. Heatmap:

A heatmap provides a visual representation of the correlation between numerical variables.

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

5. Crosstab:

Crosstab generates a contingency table, which summarizes the frequency of categories within two categorical variables.

```python
cross_tab = pd.crosstab(data['Store'], data['Holiday_Flag'])
print(cross_tab)
```

6. Line Plot:

A line plot can be used to visualize the trend of numerical data over time (for example, weekly sales data).

```python
# Assuming the 'Date' column is properly formatted as a datetime object
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Weekly_Sales'])
plt.title('Line Plot of Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

# 2.    Bivariate Analysis

1. Bar Plot (Bivariate):

A bivariate bar plot can help visualize the relationship between two categorical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot for the relationship between 'Store' and 'Holiday_Flag'
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Store', hue='Holiday_Flag')
plt.title('Bivariate Bar Plot of Store and Holiday_Flag')
plt.xticks(rotation=90)
plt.show()
```

2. Box Plot (Bivariate):

A bivariate box plot can help visualize the relationship between a numerical variable and a categorical variable.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Holiday_Flag', y='Weekly_Sales')
plt.title('Bivariate Box Plot of Holiday_Flag and Weekly Sales')
plt.show()
```

3. FacetGrid (for Bivariate Analysis):

You can use FacetGrid to create multiple plots for different categories within two variables, allowing you to explore their relationships.

```python
g = sns.FacetGrid(data, col='Holiday_Flag', row='Store')
g.map(plt.scatter, 'Temperature', 'Weekly_Sales')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Create a FacetGrid for bivariate analysis
g = sns.FacetGrid(data, col='Store', row='Holiday_Flag', margin_titles=True)

# Define the plot type in each facet
g.map(sns.boxplot, 'Weekly_Sales', color='blue')

# Set labels and titles
g.set_axis_labels("Weekly Sales", "Count")
g.set_titles(col_template="Store {col_name}", row_template="Holiday_Flag {row_name}")

# Add spacing between facets
g.tight_layout()

# Show the FacetGrid
plt.show()
```

4. Heatmap (Correlation):

A heatmap can visualize the correlation between numerical variables, showing how they are related to each other.

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

5. Crosstab (Bivariate):

Crosstab can be used to create a contingency table summarizing the relationship between two categorical variables.

```python
cross_tab = pd.crosstab(data['Store'], data['Holiday_Flag'])
print(cross_tab)
```

6. Line Plot (Bivariate):

You can create a line plot to visualize the relationship between two numerical variables, such as weekly sales over time.

```python
# Assuming the 'Date' column is properly formatted as a datetime object
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Date', y='Weekly_Sales', hue='Holiday_Flag')
plt.title('Bivariate Line Plot of Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```

# 3.    Multivariate Analysis

1. Bar Plot (Multivariate):

You can create a multivariate bar plot to visualize the distribution of a categorical variable across multiple categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot to explore the distribution of 'Store' across 'Holiday_Flag' and 'Type'
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Store', hue='Holiday_Flag',)
plt.title('Multivariate Bar Plot of Store, Holiday_Flag, and')
plt.xticks(rotation=90)
plt.show()
```

2. Box Plot (Multivariate):

A multivariate box plot can help visualize the distribution of numerical data across multiple categories.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, y='Weekly_Sales', hue='Holiday_Flag')
plt.title('Multivariate Box Plot of Type, Weekly Sales, and Holiday_Flag')
plt.show()
```

3. FacetGrid (for Multivariate Analysis):

FacetGrid can create a grid of subplots for visualizing relationships between multiple variables.

```python
g = sns.FacetGrid(data, row='Store', col='Holiday_Flag', margin_titles=True)

g.map(sns.histplot, 'Weekly_Sales', bins=20)
```

4. Heatmap (Multivariate Correlation):

A heatmap can show the correlation between multiple numerical variables in the dataset.

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Multivariate Heatmap of Correlation Matrix')
plt.show()
```

5. Crosstab (Multivariate):

You can use crosstab to create a contingency table summarizing the relationship between multiple categorical variables.

```python
cross_tab = pd.crosstab(data['Store'], [data['Holiday_Flag'], data['Store']])
print(cross_tab)
```

6. Line Plot (Multivariate):

A line plot can help visualize trends in numerical data over time, considering multiple variables, such as 'Date' and 'Weekly_Sales.'

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Date', y='Weekly_Sales', hue='Store', style='Holiday_Flag')
plt.title('Multivariate Line Plot of Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.show()
```
